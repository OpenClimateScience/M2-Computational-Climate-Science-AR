<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">دراسة الجفاف باستخدام بيانات المناخ</h2>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">قد نظنّ أننا نمتلك فهمًا حدسيًا للجفاف، لكن الجفاف يتخذ أشكالًا متعددة <span dir="ltr" style="unicode-bidi: isolate;">(Wilhite and Glantz 1985)</span>.</p>
<ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>الجفاف الأرصادي</strong> هو فترة يكون فيها الهطول (الأمطار) أقل من مقدار متوقع. ويختلف مقدار الهطول المتوقع بين الأماكن، كما يعتمد على الوقت من السنة <span dir="ltr" style="unicode-bidi: isolate;">(Palmer 1965)</span>.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>الجفاف الزراعي</strong> يحدث عندما يتجاوز طلب النبات على الماء، وبخاصة طلب المحاصيل، إمدادات الماء أيًا كان مصدرها.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>الجفاف الهيدرولوجي</strong> يصف آثار الظروف الجافة على الهيدرولوجيا السطحية أو تحت السطحية؛ أي يمكن استخدامه لوصف انخفاض التصريف النهري أو انخفاض مستويات الخزانات. وبسبب احتمال وجود تأخر زمني بين عجز الرطوبة وتغير الهيدرولوجيا، يكون الجفاف الهيدرولوجي غالبًا غير متزامن مع أنواع أخرى من الجفاف.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>الجفاف الاجتماعي-الاقتصادي</strong> يُعرَّف من حيث الآثار الاجتماعية والاقتصادية للظروف الجافة؛ مثل تغير أسعار المحاصيل أو أعلاف الحيوانات أو الكلأ؛ أو فقدان سبل العيش الزراعية أو السمكية.</li>
</ul>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ويمكن أن نضيف إلى هذه القائمة نوعًا من الجفاف حظي باعتراف أوسع مؤخرًا مع تحسن تقنيات مراقبة رطوبة التربة، وهو <strong>جفاف رطوبة التربة</strong>، أي عجز رطوبة التربة على وجه الخصوص. وقد قدمنا سابقًا مفهوم <strong>الجفاف الخاطف</strong>، وهو نوع من جفاف رطوبة التربة يتميز بسرعة ظهوره والانخفاض السريع في رطوبة التربة.</p>
</div>

<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">قياس الجفاف كميًا</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>توجد عدة مقاربات لقياس آثار الجفاف انطلاقًا من بيانات المناخ.</strong> تُستخدم <strong>مؤشرات الجفاف</strong>، مثل <span dir="ltr" style="unicode-bidi: isolate;">Palmer Drought Severity Index (PDSI)</span> <span dir="ltr" style="unicode-bidi: isolate;">(Palmer 1965)</span> و<span dir="ltr" style="unicode-bidi: isolate;">Standardized Precipitation-Evapotranspiration Index (SPEI)</span> <span dir="ltr" style="unicode-bidi: isolate;">(Vicente-Serrano et al. 2010)</span>، على نطاق واسع لأنها توفر مقياسًا بلا أبعاد لشدة الجفاف يسهل تفسيره. كما يمكن استخدام <strong>المئينات أو الترتيب</strong> لتمثيل الظروف الهيدرولوجية؛ فعلى سبيل المثال، من الشائع وصف حالة الغطاء الثلجي (و"جفاف الغطاء الثلجي") كنسبة مئوية من العمق التاريخي الوسيط للغطاء الثلجي في تاريخ معيّن.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ويمكن أيضًا اتباع مقاربة هيدرولوجية أو قائمة على ميزان الماء، لكنها تتطلب بيانات جيدة عن مكونات <strong>ميزانية مياه على مستوى الحوض</strong>. سنتحدث عن ميزانيات المياه لاحقًا. في الوقت الحالي، يمكننا تخيل "نموذج دلو" بسيط؛ يدخل الماء إلى النظام البيئي على شكل <strong>هطول</strong>، ويغادر على شكل <strong>تبخر-نتح</strong> <span dir="ltr" style="unicode-bidi: isolate;">(ET)</span> (مجموع التبخر من الأسطح الرطبة والنتح من النباتات). رياضيًا، يمكن تمثيل نموذج الدلو كما يلي.</p>
$$
\text{Available water} = \text{Precipitation} - \text{ET}
$$
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">سنستخدم <strong>التبخر-النتح المحتمل</strong> <span dir="ltr" style="unicode-bidi: isolate;">(PET)</span> بوصفه مقياسنا لـ <span dir="ltr" style="unicode-bidi: isolate;">ET</span>، إذ يمثل كمية الماء التي يمكن أن تتبخر (وتنتح) بالنظر إلى مقدار الطاقة المتاحة (وخاصة الحرارة والإشعاع الشمسي) لتبخير الماء. ويمكن تعريف الجفاف، بما ينسجم مع تعريفات الجفاف الأرصادي والزراعي والهيدرولوجي، على أنه فترة يكون فيها الهطول أقل بكثير من كمية الماء التي يمكن فقدها عبر <span dir="ltr" style="unicode-bidi: isolate;">ET</span>.</p>
</div>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تنظيم ملفات المشروع</h2>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">سنعمل اليوم على مجموعة بيانات تُسمّى <a href="https://www.chc.ucsb.edu/data/chirps" dir="ltr">Climate Hazards Group InfraRed Precipitation with Station data (CHIRPS)</a>. تجمع <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span> بين بيانات الاستشعار عن بُعد ومجموعات بيانات محطات الطقس العالمية لإنتاج سجل عالمي شبكي للهطول. ولا تُنتَج <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span> بواسطة <span dir="ltr" style="unicode-bidi: isolate;">NASA</span>، لكنها تُعد من أفضل السجلات العالمية المتاحة للهطول.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يمكن تنزيل بيانات <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span> من مصادر متعددة، لكن لا توجد واجهة مثل وصول <span dir="ltr" style="unicode-bidi: isolate;">EarthData Search</span> عبر السحابة. بدلًا من ذلك، يمكن تنزيل ملفات البيانات الفردية يدويًا من خادم عبر <a href="https://data.chc.ucsb.edu/products/CHIRPS-2.0/" dir="ltr">هذا الرابط</a>، وتشرح <a href="https://data.chc.ucsb.edu/products/CHIRPS-2.0/README-CHIRPS.txt" dir="ltr">ملف README هذا</a> كيفية تنزيل الملفات.</p>
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F6A9; انتبه</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>لأننا ما زلنا نتعلم، وبدلًا من تنزيل جميع تلك الملفات يدويًا، سنستخدم مجموعة بيانات مُحضّرة.</strong></p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F449; <strong>اضغط للتنزيل</strong>: <span dir="ltr" style="unicode-bidi: isolate;"><a href="http://files.ntsg.umt.edu/data/ScienceCore/CHIRPS-v2_Africa_monthly_2014-2024.nc" dir="ltr">CHIRPS-v2_Africa_monthly_2014-2024.nc</a></span></p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تم إنتاج هذه المجموعة عبر دمج <a href="https://data.chc.ucsb.edu/products/CHIRPS-2.0/africa_monthly/tifs/" dir="ltr">ملفات CHIRPS الشهرية الفردية لإفريقيا</a> من <span dir="ltr" style="unicode-bidi: isolate;">2014</span> حتى <span dir="ltr" style="unicode-bidi: isolate;">2023</span>. ويمكنك الاطلاع على <a href="https://github.com/OpenClimateScience/M2-Computational-Climate-Science/blob/main/scripts/20240611_process_CHIRPS_monthly_into_stack.py" dir="ltr">البرنامج النصي الذي استُخدم لدمج الملفات</a>. هذا السجل الممتد لعشر سنوات أقصر مما نفضّل عادةً لاستنتاج التذبذب المناخي، لكننا نحاول إبقاء مجموعة البيانات صغيرة.</p>
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F3AF; أفضل ممارسة</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>نبدأ تحليلًا جديدًا. لنأخذ لحظة لتنظيم نظام ملفات المشروع.</strong> انظر إلى مثال شجرة الملفات أدناه واستخدمه لإلهام تنظيمك.</p>

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تحت الصورة: مثال لشجرة ملفات منظمة لمشروع <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span>.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">مع تزايد خبرتنا في كتابة شيفرة بايثون العلمية، ينبغي أن يكون أحد أهدافنا كتابة شيفرة تعمل أيضًا بوصفها توثيقًا لسير العمل. إن أحد التحديات الأساسية في العلم المفتوح القابل لإعادة الإنتاج هو ربط النتائج العلمية بالشيفرة التي أنشأتها.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">إحدى طرق ربط النتائج ببرامجنا النصية هي اعتماد <strong>نظام تسمية ثابت</strong> بمعرّف فريد يربط الملفات معًا. ومن المقاربات استخدام تاريخ اليوم بصيغة <span dir="ltr" style="unicode-bidi: isolate;">YYYYMMDD</span> (سنة، شهر، يوم). هذا الرقم المكوّن من 8 أرقام سيكون فريدًا دائمًا لأن كل يوم يختلف عن الآخر. إذا استخدمت تاريخ اليوم في اسم ملف البرنامج النصي وأي ملفات ناتجة عنه، فستمتلك طريقة لربط تلك الملفات ببعضها.</p>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">قراءة سلسلة زمنية مناخية</h2>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>لاكتساب خبرة أكبر في استخدام <span dir="ltr" style="unicode-bidi: isolate;">xarray</span> لتحليل مجموعات بيانات المناخ، سندرس جفاف عام <span dir="ltr" style="unicode-bidi: isolate;">2024</span> في تيارت، الجزائر.</strong> وقد أدى الجفاف الممتد لأشهر في تيارت إلى اضطرابات عنيفة في المنطقة وفقًا لتقرير <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://apnews.com/article/algeria-drought-rain-tebboune-tiaret-riots-09ce23f4ba235aaf1e3afecc7bfe3574" dir="ltr">AP News</a></span>. هل يمكننا تحديد مدى اتساع وشدة جفاف تيارت باستخدام بيانات <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span>؟</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">كما في السابق، سنستخدم <span dir="ltr" style="unicode-bidi: isolate;"><code>xr.open_mfdataset()</code></span> ("فتح مجموعة بيانات متعددة الملفات") لفتح هذه المجموعة. وعلى الرغم من وجود ملف واحد فقط، فإن <span dir="ltr" style="unicode-bidi: isolate;"><code>xr.open_mfdataset()</code></span> تتيح لنا الوصول إلى ميزات مفيدة في مكتبة <span dir="ltr" style="unicode-bidi: isolate;">dask</span>.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">على سبيل المثال، لاحظ أن المخرجات المرتبطة بـ <span dir="ltr" style="unicode-bidi: isolate;"><code>"precip"</code></span> تتضمن معلومات عن الحجم الكلي للمصفوفة <span dir="ltr" style="unicode-bidi: isolate;">(1.12 GiB)</span>.</p>
</div>


```python
import xarray as xr
import numpy as np

ds = xr.open_mfdataset('data_raw/CHIRPS/CHIRPS-v2_Africa_monthly_2014-2024.nc')
ds['precip']
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">لم تُحمَّل هذه <span dir="ltr" style="unicode-bidi: isolate;">1.12</span> غيغابايت إلى الذاكرة بعد؛ بل يشير المتغير <span dir="ltr" style="unicode-bidi: isolate;"><code>ds</code></span> إلى <strong>تمثيل</strong> لمجموعة البيانات المخزنة على القرص الصلب. وهذا مثال آخر على <strong>التقييم الكسول</strong> <span dir="ltr" style="unicode-bidi: isolate;">(lazy evaluation)</span>، وهو أمر مفيد. لن تقرأ <span dir="ltr" style="unicode-bidi: isolate;">xarray</span> البيانات إلى الذاكرة إلا عندما نكون مستعدين لتنفيذ حسابات. وبما أننا استخدمنا <span dir="ltr" style="unicode-bidi: isolate;"><code>xr.open_mfdataset()</code></span>، فستضمن <span dir="ltr" style="unicode-bidi: isolate;">xarray</span> و<span dir="ltr" style="unicode-bidi: isolate;">dask</span> ألا يتجاوز حجم البيانات المحمّلة ذاكرة الحاسوب المتاحة؛ وإذا كانت مجموعة البيانات أكبر من الذاكرة، فسيتم تحميل أجزاء أصغر ومعالجتها تباعًا.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>الخطوة التالية هي استبدال قيم <span dir="ltr" style="unicode-bidi: isolate;">NoData</span> في مجموعة البيانات.</strong> تعمل الدالة <span dir="ltr" style="unicode-bidi: isolate;"><code>xr.where()</code></span> تمامًا مثل <span dir="ltr" style="unicode-bidi: isolate;"><code>np.where()</code></span>. نقدّم تعبيرًا شرطيًا على <span dir="ltr" style="unicode-bidi: isolate;"><code>xarray.DataArray</code></span> كوسيط أول، ثم القيمة التي تُعاد إذا كان الشرط صحيحًا، ثم القيمة التي تُعاد إذا كان الشرط خاطئًا.</p>
</div>


```python
ds['precip'] = xr.where(ds['precip'] == -9999, np.nan, ds['precip'])
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">أصبحنا الآن جاهزين لرسم بعض البيانات. يمكننا اختيار تاريخ معيّن باستخدام الطريقة <span dir="ltr" style="unicode-bidi: isolate;"><code>sel()</code></span> عبر تحديد أننا نريد اقتطاع بُعد <span dir="ltr" style="unicode-bidi: isolate;"><code>time</code></span>. ويمكن اقتطاع أي بُعد يحتوي على إحداثيات <span dir="ltr" style="unicode-bidi: isolate;">datetime64[ns]</span> باستخدام طابع زمني منسّق وفق معيار <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://en.wikipedia.org/wiki/ISO_8601" dir="ltr">ISO 8601</a></span>، مثل <span dir="ltr" style="unicode-bidi: isolate;"><code>"2023-07-01"</code></span>.</p>
</div>


```python
ds['precip'].sel(time = '2023-07-01').plot()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F3C1; تحدٍّ: حساب متوسط الهطول السنوي</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تتضمن مجموعة البيانات هطولًا شهريًا على مدى عشر سنوات. أنشئ رسمًا لمتوسط الهطول السنوي. <strong>ينبغي تنفيذ ذلك على خطوتين.</strong></p>
<ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">استخدم الطريقة <span dir="ltr" style="unicode-bidi: isolate;"><code>resample()</code></span> ثم <span dir="ltr" style="unicode-bidi: isolate;"><code>sum()</code></span> لحساب <strong>إجمالي الهطول</strong> في كل سنة. يمكنك الرجوع إلى <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects" dir="ltr">هذا الجدول</a></span> لمعرفة ترددات إعادة التجميع المتاحة.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ثم احسب متوسط الهطول السنوي.</li>
</ul>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">وسّع الخلية أدناه للاطلاع على الحل.</p>
</div>


```python
# First, add up monthly precipitation (over 12 months) in each year
annual_precip = ds['precip'].resample(time = 'YS').sum()
# Then, calculate the average (mean) annual precipitation
mean_annual_precip = annual_precip.mean('time')
mean_annual_precip.plot()
```

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">حساب المتوسطات الموسمية والمتدحرجة</h2>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">بالإضافة إلى إعادة تجميع بياناتنا إلى فواصل سنوية، يمكننا إعادة تجميع البيانات لتعكس متوسطات موسمية. لتحقيق ذلك نستخدم تردد <span dir="ltr" style="unicode-bidi: isolate;"><code>"QS"</code></span> (بداية الربع). تذكّر أن <span dir="ltr" style="unicode-bidi: isolate;"><code>resample()</code></span> يجب أن تتبعها دالة تجميع. نستخدم <span dir="ltr" style="unicode-bidi: isolate;"><code>sum()</code></span> لأننا نريد جمع الهطول الشهري ضمن كل موسم.</p>
</div>


```python
seasonal_precip = ds['precip'].resample(time = 'QS').sum()
seasonal_precip
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F449; <strong>ملاحظة</strong> إن الأرباع تُمثَّل بشهر البداية. الربع الأول <span dir="ltr" style="unicode-bidi: isolate;">(Q1)</span> يبدأ في يناير، والربع الثاني <span dir="ltr" style="unicode-bidi: isolate;">(Q2)</span> يبدأ في أبريل، وهكذا.</p>
</div>


```python
seasonal_precip.coords['time']
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">منهجية Split-Apply-Combine</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">استخدمنا أعلاه الدالة <span dir="ltr" style="unicode-bidi: isolate;"><code>resample()</code></span> لتجميع البيانات الشهرية إلى بيانات موسمية (أو ربع سنوية). وطريقة أخرى لتجميع البيانات هي حساب المتوسط الموسمي الكلي؛ مثلًا، ما متوسط الهطول الموسمي في الربع الأول (من يناير إلى أبريل)؟</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">للإجابة، يمكننا أخذ بياناتنا الموسمية و<strong>تجميعها حسب الموسم ثم حساب المتوسط داخل كل موسم</strong>. تذكّر أن <span dir="ltr" style="unicode-bidi: isolate;"><code>xarray.DataArray</code></span> يحتوي على الطريقة <span dir="ltr" style="unicode-bidi: isolate;"><code>groupby()</code></span> لهذا الغرض.</p>
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F6A9; انتبه</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>في المثال أدناه، نقوم بالتجميع حسب <span dir="ltr" style="unicode-bidi: isolate;"><code>"time.month"</code></span> رغم أننا لا نمتلك بيانات شهرية. لدينا بيانات موسمية.</strong> إحداثيات <span dir="ltr" style="unicode-bidi: isolate;"><code>time</code></span> في البيانات الموسمية ما تزال تتضمن مكوّن <span dir="ltr" style="unicode-bidi: isolate;">month</span>، ولا توجد إلا أربعة أشهر فريدة: يناير، أبريل، يوليو، وأكتوبر، وهي تمثل بداية كل ربع (أو موسم). لذلك، عندما نُجمّع حسب الشهر فإننا في الواقع نُجمّع حسب الموسم.</p>
</div>


```python
mean_seasonal_precip = seasonal_precip.groupby('time.month').mean()
mean_seasonal_precip
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">عندما نستخدم <span dir="ltr" style="unicode-bidi: isolate;"><code>groupby()</code></span> مع دالة تجميع مثل <span dir="ltr" style="unicode-bidi: isolate;"><code>mean()</code></span> فإننا نطبق منهجية <strong>Split-Apply-Combine</strong>.</p>
<ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">نقوم <strong>بتقسيم البيانات</strong> إلى مجموعات مختلفة. في هذا المثال، المجموعات هي الأشهر الفريدة (المواسم).</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ثم <strong>نطبّق</strong> دالة على كل مجموعة. في هذا المثال، طبّقنا <span dir="ltr" style="unicode-bidi: isolate;"><code>mean()</code></span> على كل مجموعة.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">وأخيرًا <strong>نُجمّع النتائج</strong> في مخرَج واحد. تتم هذه الخطوة تلقائيًا وتنتج مكعب بيانات واحد لكل المجموعات. في هذا المثال، نحصل على مكعب بيانات له بُعد جديد باسم <span dir="ltr" style="unicode-bidi: isolate;"><code>month</code></span> مع 4 عناصر.</li>
</ul>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>يمكننا الآن رسم متوسط الهطول الموسمي في الربع الأول</strong> (من يناير إلى أبريل).</p>
</div>


```python
mean_seasonal_precip.sel(month = 1).plot()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">المتوسطات المتدحرجة</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">لنركز الآن على منطقة تيارت. مرة أخرى، يمكننا استخدام <span dir="ltr" style="unicode-bidi: isolate;"><code>sel()</code></span> والدالة المدمجة <span dir="ltr" style="unicode-bidi: isolate;"><code>slice()</code></span> لإنشاء اقتطاع مكاني حول تيارت، تقريبًا من <span dir="ltr" style="unicode-bidi: isolate;">0.8</span> إلى <span dir="ltr" style="unicode-bidi: isolate;">1.8</span> درجة شرقًا، ومن <span dir="ltr" style="unicode-bidi: isolate;">35.1</span> إلى <span dir="ltr" style="unicode-bidi: isolate;">36.1</span> درجة شمالًا.</p>
<h3 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F6A9; انتبه</h3>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>لاحظ أننا في المثال أدناه نُعطي إحداثيات خط العرض (أو <span dir="ltr" style="unicode-bidi: isolate;">y</span>) بترتيب تنازلي</strong> <span dir="ltr" style="unicode-bidi: isolate;"><code>slice(36.1, 35.1)</code></span>. ذلك لأن قيم البداية والنهاية في <span dir="ltr" style="unicode-bidi: isolate;"><code>slice()</code></span> يجب أن تكون دائمًا بنفس ترتيب إحداثيات البيانات. هذا الملف مرتّب من الشمال إلى الجنوب (من <span dir="ltr" style="unicode-bidi: isolate;">+90</span> إلى <span dir="ltr" style="unicode-bidi: isolate;">-90</span>).</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">لنحسب أيضًا الهطول السنوي في منطقة تيارت في كل سنة باستخدام <span dir="ltr" style="unicode-bidi: isolate;"><code>resample()</code></span> و<span dir="ltr" style="unicode-bidi: isolate;"><code>sum()</code></span> كما فعلنا سابقًا.</p>
</div>


```python
# Note the slice() order for the y coordinate must go from +90 to -90
ds_tiaret = ds.sel(x = slice(0.8, 1.8), y = slice(36.1, 35.1))

tiaret_precip = ds_tiaret['precip'].resample(time = 'YS').sum()
tiaret_precip
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يمكننا بعد ذلك حساب <strong>شذوذ الهطول السنوي</strong> عبر طرح متوسط الهطول السنوي العام.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ولعرض السلسلة الزمنية كرسم خطي، يجب أيضًا أن نأخذ المتوسط عبر منطقة تيارت، أي المنطقة ذات <span dir="ltr" style="unicode-bidi: isolate;">20</span> × <span dir="ltr" style="unicode-bidi: isolate;">20</span> بكسل التي اخترناها باستخدام <span dir="ltr" style="unicode-bidi: isolate;"><code>slice()</code></span> أعلاه.</p>
</div>


```python
tiaret_anomaly = tiaret_precip - tiaret_precip.mean('time')

# Average over the 20-by-20 pixel region around Tiaret
tiaret_anomaly.mean(['x', 'y']).plot()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يبدو عام <span dir="ltr" style="unicode-bidi: isolate;">2024</span> سلبيًا جدًا في الرسم أعلاه، لكن لا ننسى أن مجموعة البيانات تتضمن أول خمسة أشهر فقط من <span dir="ltr" style="unicode-bidi: isolate;">2024</span>. وبما أننا نقارن الهطول السنوي في كل سنة، فلا يصح استخدام بيانات جزئية من <span dir="ltr" style="unicode-bidi: isolate;">2024</span>. <strong>دعنا ننظر بدلًا من ذلك إلى شذوذات الهطول الشهرية.</strong></p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">مرة أخرى، نستخدم منهجية <strong>Split-Apply-Combine</strong> لتقسيم البيانات إلى مجموعات شهرية، ثم تطبيق دالة لحساب الشذوذ. لاحظ أن الدالة المستخدمة هي دالة <span dir="ltr" style="unicode-bidi: isolate;">lambda</span>. تعمل على كمية <span dir="ltr" style="unicode-bidi: isolate;">x</span> عبر طرح متوسطها <span dir="ltr" style="unicode-bidi: isolate;"><code>x.mean()</code></span> من كل خطوة زمنية شهرية. ولأننا جمعنا حسب <span dir="ltr" style="unicode-bidi: isolate;"><code>"time.month"</code></span>، فإن شذوذات الهطول تكون بالنسبة للشهر التقويمي الموافق. على سبيل المثال، شذوذ <span dir="ltr" style="unicode-bidi: isolate;">2014-01-01</span> يُقاس بالنسبة لمتوسط شهر يناير العام.</p>
</div>


```python
precip_anomaly = ds_tiaret['precip'].groupby('time.month').apply(lambda x: x - x.mean())
precip_anomaly
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يمكننا الآن رسم شذوذات الهطول الشهرية.</p>
</div>


```python
# Again, we average over the spatial domain of our Tiaret subset
precip_anomaly.mean(['x', 'y']).plot()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يوجد قدر كبير من التذبذب في هذا الرسم يجعل من الصعب رؤية أنماط واضحة. <strong>يمكننا تطبيق متوسط متدحرج باستخدام الطريقة <span dir="ltr" style="unicode-bidi: isolate;"><code>rolling()</code></span> في <span dir="ltr" style="unicode-bidi: isolate;"><code>xarray.DataArray</code></span>.</strong></p>
</div>


```python
precip_anomaly.rolling(time = 6).mean().mean(['x', 'y']).plot()
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;"><strong>يؤدي ذلك إلى تنعيم البيانات، ما يساعدنا على ملاحظة أن جفاف تيارت يُعزى جزئيًا إلى اتجاه تجفّف متعدد السنوات.</strong></p>
</div>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">نموذج دلو بسيط</h2>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تمنحنا شذوذات الهطول التي حسبناها معلومات حول تغير إمدادات الماء (الهطول)، لكنها لا تمنحنا معلومات حول تغير الطلب أو الخسائر، مثل أثر غلاف جوي أدفأ أو أكثر جفافًا. تذكّر أن مقياسًا أفضل لتأثيرات الجفاف الهيدرولوجي هو <strong>الفرق</strong> بين الإمداد والطلب كما ينعكس في الهطول ناقص <span dir="ltr" style="unicode-bidi: isolate;">PET</span>.</p>
$$
\text{Available water} \approx \text{Precipitation} - \text{PET}
$$
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">سننظر هنا في تباين بسيط على هذا المؤشر. وهو <strong>نسبة الهطول إلى <span dir="ltr" style="unicode-bidi: isolate;">PET</span></strong>، والتي يمكن تفسيرها على أنها نسبة (أو مئوية) من الطلب المائي (<span dir="ltr" style="unicode-bidi: isolate;">PET</span>) يتم تعويضها عبر الهطول.</p>
$$
\text{Percentage replenished} \approx 100\times \frac{\text{Precipitation}}{\text{PET}}
$$
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">رغم وجود <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://doi.org/10.1029/2022WR033447" dir="ltr">طرق عديدة لحساب PET</a></span> <span dir="ltr" style="unicode-bidi: isolate;">(Pimentel et al. 2023)</span>، سنستخدم <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://www.fao.org/4/X0490E/x0490e07.htm#minimum%20data%20requirements" dir="ltr">the Hargreaves method</a></span> <span dir="ltr" style="unicode-bidi: isolate;">(Allen et al. 2000)</span> لأنها تتطلب بيانات حرارة فقط.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">ستأتي بيانات الحرارة التي سنستخدمها من <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://climatedataguide.ucar.edu/climate-data/terraclimate-global-high-resolution-gridded-temperature-precipitation-and-other-water" dir="ltr">TerraClimate</a></span> <span dir="ltr" style="unicode-bidi: isolate;">(Abatzoglou et al. 2018)</span>، وهي مجموعة بيانات عالمية شبكية عالية الدقة لدرجة الحرارة. توجد طرق عديدة لتنزيل بيانات <span dir="ltr" style="unicode-bidi: isolate;">TerraClimate</span>، وبسبب دقتها العالية قد تكون أحجام الملفات كبيرة. لذلك سنستخدم ملفًا مُنشأً من <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://climate.northwestknowledge.net/NWTOOLBOX/formattedDownloads.php" dir="ltr">هذه الأداة الإلكترونية للاقتطاع</a></span>.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">&#x1F449; <strong>اضغط للتنزيل</strong>: <span dir="ltr" style="unicode-bidi: isolate;"><a href="http://files.ntsg.umt.edu/data/ScienceCore/terraclimate_35.3709N_-1.3218W.csv" dir="ltr">terraclimate_35.3709N_-1.3218W.csv</a></span></p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يحتوي هذا الملف على بيانات سلسلة زمنية مناخية لمنطقة تيارت من <span dir="ltr" style="unicode-bidi: isolate;">1958</span> حتى الوقت الحاضر.</p>
</div>


```python
import pandas as pd

pet = pd.read_csv('data_raw/terraclimate_35.3709N_-1.3218W.csv', skiprows = 11)

# Let's look at just the data since 2014
pet = pet[pet['Year'] >= 2014]
pet
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تتضمن مجموعة بيانات <span dir="ltr" style="unicode-bidi: isolate;">TerraClimate</span> بيانات <span dir="ltr" style="unicode-bidi: isolate;">PET</span> محسوبة بوصفها "reference ET" وفق نهج <span dir="ltr" style="unicode-bidi: isolate;">FAO Penman-Monteith</span>. في الدرس التالي سنحسب <span dir="ltr" style="unicode-bidi: isolate;">PET</span> بأنفسنا باستخدام طريقة <span dir="ltr" style="unicode-bidi: isolate;">Hargreaves</span>.</p>
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">لكن في الوقت الحالي، دعنا نجمع تقدير <span dir="ltr" style="unicode-bidi: isolate;">PET</span> من <span dir="ltr" style="unicode-bidi: isolate;">TerraClimate</span> مع بيانات هطول <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span> لحساب مؤشر هيدرولوجي بسيط للجفاف.</p>
</div>


```python
ds_tiaret['precip']
```

<div dir="rtl" style="direction: rtl; unicode-bidi: isolate; text-align: right;">
<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">نختار أول <span dir="ltr" style="unicode-bidi: isolate;">120</span> شهرًا من هطول <span dir="ltr" style="unicode-bidi: isolate;">CHIRPS</span>، ونحسب المتوسط عبر منطقة تيارت، ثم نقسم ذلك على <span dir="ltr" style="unicode-bidi: isolate;">TerraClimate PET</span>.</p>
</div>


```python
from matplotlib import pyplot

# NOTE: The CHIRPS data extend monthly through May 2024, but the PET data
#    do not, so we have to subset the CHIRPS data to the first 120 months
precip_pet_ratio = 100 * ds_tiaret['precip'].isel(time = slice(0, 120)).mean(['x','y']) / pet['pet(mm)']
precip_pet_ratio.plot()
pyplot.ylabel('Precipitation-to-PET Ratio (%)')
pyplot.title('Precipitation-to-PET Ratio for Tiaret, Algeria')
pyplot.savefig('results/20240610_Tiaret_precip-to-PET_ratio.png', dpi = 172)
```

<p style="direction: rtl; unicode-bidi: plaintext; text-align: right;">وهنا نرى مرة أخرى أن جفاف تيارت الحديث تسبقه عدة سنوات من عدم كفاية هطول الموسم الرطب. فبينما توفر معظم المواسم الرطبة أكثر من <span dir="ltr" style="unicode-bidi: isolate;">100%</span> من الماء الذي يتطلبه <span dir="ltr" style="unicode-bidi: isolate;">PET</span>، وفّرت المواسم الرطبة الأخيرة هطولًا أقل مما قد يُفقد عبر <span dir="ltr" style="unicode-bidi: isolate;">PET</span>.</p>
</div>

<h2 style="direction: rtl; unicode-bidi: plaintext; text-align: right;">الخلاصة</h2>
<ul>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">توجد أنواع مختلفة من الجفاف، كما توجد طرق مختلفة لقياسه اعتمادًا على بيانات المناخ. إضافةً إلى مؤشرات مثل <span dir="ltr" style="unicode-bidi: isolate;">PDSI</span> و<span dir="ltr" style="unicode-bidi: isolate;">SPEI</span>، يمكن استخدام مقاربة ميزان ماء بسيطة. في هذه الحالة تُوصَف ظروف الجفاف بعجز كبير في إمداد الماء، ويمكن حساب ذلك كفرق بين الهطول و<span dir="ltr" style="unicode-bidi: isolate;">PET</span>.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يساعد <strong>نظام تسمية ثابت للملفات</strong> بمعرّف فريد مثل تاريخ اليوم بصيغة <span dir="ltr" style="unicode-bidi: isolate;">YYYYMMDD</span> على ربط ملفات الشيفرة، والمجموعات المشتقة، والأشكال أو الجداول الناتجة. هذا يسهل معرفة أي برنامج نصي أنشأ مخرَجًا معينًا.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يمكن استخدام <span dir="ltr" style="unicode-bidi: isolate;"><code>xarray.open_mfdataset()</code></span> لفتح ملف أو أكثر مع <span dir="ltr" style="unicode-bidi: isolate;">dask</span>، ما يسمح بتحليل مجموعات كبيرة حتى لو لم تلائم الذاكرة.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">يمكن استخدام منهجية <strong>Split-Apply-Combine</strong> لتقسيم البيانات إلى مجموعات، وتطبيق دالة على كل مجموعة، ثم دمج النتائج. في <span dir="ltr" style="unicode-bidi: isolate;">xarray</span> يتحقق ذلك عبر <span dir="ltr" style="unicode-bidi: isolate;"><code>groupby()</code></span> ثم دوال مثل <span dir="ltr" style="unicode-bidi: isolate;"><code>mean()</code></span> أو <span dir="ltr" style="unicode-bidi: isolate;"><code>sum()</code></span> أو <span dir="ltr" style="unicode-bidi: isolate;"><code>apply()</code></span>. يمكن الرجوع إلى <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://xarray.pydata.org/en/v2024.06.0/user-guide/groupby.html" dir="ltr">xarray groupby documentation</a></span>.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تجمع مجموعة بيانات <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://www.chc.ucsb.edu/data/chirps" dir="ltr">CHIRPS</a></span> بين الاستشعار عن بُعد وبيانات محطات الطقس لإنتاج سجل عالمي شبكي للهطول.</li>
<li style="direction: rtl; unicode-bidi: plaintext; text-align: right;">تُعد <span dir="ltr" style="unicode-bidi: isolate;"><a href="https://climatedataguide.ucar.edu/climate-data/terraclimate-global-high-resolution-gridded-temperature-precipitation-and-other-water" dir="ltr">TerraClimate</a></span> <span dir="ltr" style="unicode-bidi: isolate;">(Abatzoglou et al. 2018)</span> مجموعة بيانات عالمية عالية الدقة تتضمن درجة الحرارة والإشعاع الشمسي و<span dir="ltr" style="unicode-bidi: isolate;">PET</span>.</li>
</ul>
